```python
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns  
import scipy.stats as stats 
from scipy.stats import bernoulli
```

### PMF -> Probability Mass Function

Used for discrete random variables (values are countable)

A Probability Mass Function assigns a probability to each exact value of a random variable.

PMF Definition : $p_X(x) = P(X=x)$


Key Properties: 
- $ 0 \le p_X(x) \le1 $ 

- $ \sum_x P_X(x) = 1 $

- $ P(X=x) > 0 $


Expected value or mean: $ E[X] = \sum_x P_X(x) $


Variance: $ var(X) = E[(X - E[X])^2] $

```python
p = 0.5 # Probability of Heads
x = [0,1] # 0 = Tails, 1 = Heads
pmf_values = bernoulli.pmf(x, p)
```

```python
plt.bar(x, pmf_values, tick_label=['Tails (0)', 'Heads (1)'])
plt.title('PMF of a Fair Coin Toss')
plt.ylabel('Probability')
plt.xlabel('Outcome')
plt.grid(axis='y', linestyle='--')
plt.show()
```
